```python
# Importando bibliotecas
import time
from selenium import webdriver
import pandas as pd
from time import sleep
from selenium.webdriver.common.by import By
from datetime import datetime
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Definindo driver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.maximize_window()
```

```python
url = 'https://www.linkedin.com/jobs/cientista-de-dados-vagas/?originalSubdomain=br'
driver.get(url)
time.sleep(2)

vagas = driver.find_elements(By.CLASS_NAME, 'base-card')

lista_descricao = []
results = []
while True:
    for vaga in vagas[len(lista_descricao):]:
        try:
            vaga.click() #Clicar na descrição
            sleep(2)

        
            try:
                    title = vaga.find_element(By.CLASS_NAME, 'base-search-card__title').text
            except:
                    title = None

            try:
                    subtitle = vaga.find_element(By.CLASS_NAME,'base-search-card__subtitle').text
            except:
                    subtitle = None

            try:
                    location = vaga.find_element(By.CLASS_NAME,'job-search-card__location').text
            except:
                    location = None

            try:
                    description = driver.find_element(By.CLASS_NAME,'description__text').text
            except:
                    description = None

            try:  
                    job = driver.find_element(By.CLASS_NAME,'description__job-criteria-list').text
            except:
                    job = None


            print(subtitle)

            #Criando um campo com a data de execução do Crawler
            crawled_at = datetime.now().strftime("%Y-%m-%d %H:%M")
                #Criando um dicionário de resultados
            results.append({"crawled_at": crawled_at,
                            "Cargo": title,
                            "Empresa": subtitle,
                            "Localizacao": location,
                            "Descricao": str(description).replace(" ", "_").replace("\n", " "),
                            "Criterios": str(job).replace(" ", "_").replace("\n", " ")
                                })

            sleep(2)
            
            lista_descricao.append(driver.find_element_by_class_name('description').text)
            print(len(lista_descricao))
            
        except:
            break

                  
    try:
            #Capturando a caixa com uma lista de 25 em 25 vagas      
            vagas = driver.find_elements_by_class_name('base-card')
    except:
            #Capturando a caixa com uma lista de 25 em 25 vagas   
            driver.find_element_by_xpath('//*[@id="main-content"]/section[2]/button').click()
            vagas = driver.find_elements_by_class_name('base-card')
    if len(lista_descricao) == len(vagas):
        break

# Transformando dicionario em um Dataframe
df = pd.DataFrame(results)

#Fechando o google Chrome
driver.close
```

```python
# Salvando Dataframe
df.to_excel('vagas_ds_linkedin.xlsx', encoding='utf-8-sig')


df
```
